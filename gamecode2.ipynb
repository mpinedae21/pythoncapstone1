```python
import time
import random

# Lista para almacenar las pistas encontradas a lo largo del juego.
clues = []

# Lista de eventos aleatorios para aumentar la rejugabilidad
random_events = ["You notice a black cat dart across the crime scene, knocking over a hidden piece of evidence.",
    "A strange noise catches your attention, leading you to a hidden compartment in the wall of the suspect's office.",
    "A mysterious stranger approaches you, offering a cryptic tip related to the case."]

def win():
    """Se llama cuando el jugador gana el juego."""
    print("\nCongratulations! You've caught the Zodiac Killer and solved the case.")
    print("San Francisco can rest easy once again. Thank you for playing!")
    print("""

  _______ _                 _
 |__   __| |               | |
    | |  | |__   __ _ _ __ | | __   _   _  ___  _   _
    | |  | '_ \ / _` | '_ \| |/ /  | | | |/ _ \| | | |
    | |  | | | | (_| | | | |   <   | |_| | (_) | |_| |
    |_|  |_| |_|\__,_|_| |_|_|\_\   \__, |\___/ \__,_|
                                     __/ |
                                    |___/              """)
    exit()

def fail():
    """Se llama cuando el jugador pierde el juego."""
    print("""
       _____                         ____
      / ____|                       / __ \
     | |  __  __ _ _ __ ___   ___  | |  | |_   _____ _ __
     | | |_ |/ _` | '_ ` _ \ / _ \ | |  | \ \ / / _ \ '__|
     | |__| | (_| | | | | | |  __/ | |__| |\ V /  __/ |
      \_____|\__,_|_| |_| |_|\___|  \____/  \_/ \___|_|
""")
    print("¡Oh no! El asesino ha logrado escapar por esta vez.")
    print("Pero no te rindas, sé que puedes atraparlo.")
     # Bucle while para volver a iniciar el juego
    play_again = "y"
    while play_again == "y":
      play_again = input("¿Jugar de nuevo? (y/n)")
      if play_again == "y":
        game_start() # Reinica el juego
    else:
      print("""
  _____ _              _           __           ___ _           _
 |_   _| |_  __ _ _ _ | |__ ___   / _|___ _ _  | _ \ |__ _ _  _(_)_ _  __ _
   | | | ' \/ _` | ' \| / /(_-<  |  _/ _ \ '_| |  _/ / _` | || | | ' \/ _` |
   |_| |_||_\__,_|_||_|_\_\/__/  |_| \___/_|   |_| |_\__,_|\_, |_|_||_\__, |
                                                           |__/       |___/""")
      exit()

def random_event():
    """Genera un evento aleatorio para aumentar la rejugabilidad del juego"""
    event = random.choice(random_events)
    print("\nEvent: "+ event)

def game_start():
    """Inicio del juego"""
    print("""
 __          __  _
 \ \        / / | |
  \ \  /\  / /__| | ___ ___  _ __ ___   ___
   \ \/  \/ / _ \ |/ __/ _ \| '_ ` _ \ / _ \
    \  /\  /  __/ | (_| (_) | | | | | |  __/
     \/  \/ \___|_|\___\___/|_| |_| |_|\___|
""")
    print(f"""
{'#' * 100}

INSTRUCTIONS:
Welcome to the Zodiac Killer Game
Welcome to the dark and mysterious world of Zodiac Killer. In this interactive game, you will immerse yourself in the shoes of a brave detective whose mission is to capture the infamous killer
who terrorized the San Francisco Bay Area in the 1960s and 1970s.Your task won't be easy. You'll have to make careful decisions, collect clues and solve puzzles to follow the trail of this cunning criminal.
Each choice you make can take you one step closer to your objective... or divert you from the right path.

Good Luck Detective !

{'#' * 100}\n""")
    input("Press < Enter > to begin.")
    time.sleep(2)
    stage_1()

def stage_1():
    """La primera etapa del juego."""
    print("""
     _______.___________.    ___       _______  _______     __          ______ .______       __  .___  ___.  _______         _______.  ______  _______ .__   __.  _______
    /       |           |   /   \     /  _____||   ____|   /_ |  _     /      ||   _  \     |  | |   \/   | |   ____|       /       | /      ||   ____||  \ |  | |   ____|
   |   (----`---|  |----`  /  ^  \   |  |  __  |  |__       | | (_)   |  ,----'|  |_)  |    |  | |  \  /  | |  |__         |   (----`|  ,----'|  |__   |   \|  | |  |__
    \   \       |  |      /  /_\  \  |  | |_ | |   __|      | |       |  |     |      /     |  | |  |\/|  | |   __|         \   \    |  |     |   __|  |  . `  | |   __|
.----)   |      |  |     /  _____  \ |  |__| | |  |____     | |  _    |  `----.|  |\  \----.|  | |  |  |  | |  |____    .----)   |   |  `----.|  |____ |  |\   | |  |____
|_______/       |__|    /__/     \__\ \______| |_______|    |_| (_)    \______|| _| `._____||__| |__|  |__| |_______|   |_______/     \______||_______||__| \__| |_______|

    """)
    choice = input("You find a coded message next to the victim. Do you want to examine the message or search for physical evidence? (message/evidence): ").lower()
    if choice == "message":
        random_event()  # Trigger a random event
        stage_2()  # Move to the next stage
    elif choice == "evidence":
        clues.append("physical evidence")  # Add a clue to the list
        random_event()  # Trigger a random event
        stage_2()  # Move to the next stage
    else:
        print("\nYou must choose a valid action.")
        fail() # End the game if the player doesn't choose a valid action


def stage_2():
    """La segunda etapa del juego."""
    print("""
     _______.___________.    ___       _______  _______     ___         .______     ______    __       __    ______  _______         _______.___________.    ___   .___________. __    ______   .__   __.
    /       |           |   /   \     /  _____||   ____|   |__ \   _    |   _  \   /  __  \  |  |     |  |  /      ||   ____|       /       |           |   /   \  |           ||  |  /  __  \  |  \ |  |
   |   (----`---|  |----`  /  ^  \   |  |  __  |  |__         ) | (_)   |  |_)  | |  |  |  | |  |     |  | |  ,----'|  |__         |   (----`---|  |----`  /  ^  \ `---|  |----`|  | |  |  |  | |   \|  |
    \   \       |  |      /  /_\  \  |  | |_ | |   __|       / /        |   ___/  |  |  |  | |  |     |  | |  |     |   __|         \   \       |  |      /  /_\  \    |  |     |  | |  |  |  | |  . `  |
.----)   |      |  |     /  _____  \ |  |__| | |  |____     / /_   _    |  |      |  `--'  | |  `----.|  | |  `----.|  |____    .----)   |      |  |     /  _____  \   |  |     |  | |  `--'  | |  |\   |
|_______/       |__|    /__/     \__\ \______| |_______|   |____| (_)   | _|       \______/  |_______||__|  \______||_______|   |_______/       |__|    /__/     \__\  |__|     |__|  \______/  |__| \__|
                                                                                                                                                                                                                                                                                           """)
    print("The cryptic message seems to be a riddle that points to the next target location.")
    print("You ponder over the riddle...")
    time.sleep(2)
    print("""
--------------------------------------------------------------------------------
To the north, where the giants stand tall and the sleepless lights flicker,
there lies my mark.

The wicked will perish where the money flows and the water runs deep.

Seek me where the metal birds cannot reach, and the air is filled with ambition.
--------------------------------------------------------------------------------
""")

    # Bucle while para volver a intentar si falla
    tries = 0
    location_dict = {"1": "skyscraper", "2": "docks", "3": "district"}

    while tries < 3:
        location = input("Elige ubicación (1: Skyscraper, 2: Docks, 3: District): ")
        if location in location_dict:
            if location_dict[location] == "skyscraper":
                # Si elige la ubicación correcta, procede a la siguiente etapa
                stage_3()
                return
            else:
                print("Incorrecto. Intenta de nuevo.")
        else:
            print("Debes elegir una ubicación válida.")

        # Proporciona una pista al usuario después de una elección incorrecta
        tries += 1  # Incrementa los intentos después de cada elección incorrecta
        if tries == 1:
            print("Pista: El lugar que buscas está en las alturas.")
        elif tries == 2:
            print("Pista: No está cerca del agua, ni es un área común.")

    # Si después de 3 intentos no elige la ubicación correcta, falla.
    if tries == 3:
        print("No adivinaste correctamente. Game over.")
        fail()



def stage_3():
    """The third stage of the game."""

    print("""
  ___ _                   ____  _   _____ _          __   ___    _   _
 / __| |_ __ _ __ _ ___  |__ / (_) |_   _| |_  ___   \ \ / (_)__| |_(_)_ __
 \__ \  _/ _` / _` / -_)  |_ \  _    | | | ' \/ -_)   \ V /| / _|  _| | '  \
 |___/\__\__,_\__, \___| |___/ (_)   |_| |_||_\___|    \_/ |_\__|\__|_|_|_|_|
              |___/                                                          """)

    choice = input("Do you want to try to defuse the bomb or call for help? (defuse/call): ").lower()  # Asks the player to make a decision

    if choice == "defuse":
        print("The bomb has three wires: one red (warm), one blue (cool), and one green (neutral).")
        print("Next to the bomb, you find a note with one of the Zodiac Killer's cryptic messages: 'Calm can be as cold as the vast night sky.'")
        tries = 0
        while tries < 2:
            try:
                wires_cut = int(input("Which wire do you cut? (1 for blue, 2 for red, 3 for green): "))

                if wires_cut == 1:
                    print("You cut the blue wire. The ticking stops. You've defused the bomb!")  # Code for cutting the blue wire
                    stage_4()
                    return

                elif wires_cut in [2, 3]:
                    print("You cut the wrong wire! The ticking accelerates as you're almost out of time.")  # Code for cutting the red or green wire
                    if tries == 0:
                        print("Hint: With the clock ticking, try to think calmly and decode the Zodiac's celestial clue.")
                    tries += 1

                else:
                    print("Invalid number. You panic and do nothing.")
                    tries += 1

            except ValueError:
                print("Invalid input, please enter a number.")
                tries += 1
        fail()  # Call the fail() function if the player makes two mistakes

    elif choice == "call":
        print("You decide to call for help. But the bomb squad is too far away to arrive in time...")
        fail()
    else:
        print("Overwhelmed, you hesitate too long. The ticking stops as the timer expires.")
        fail()



def stage_4():
    """The fourth and final stage of the game."""
    print("""
.___________. __    __   _______     _______  __  .__   __.      ___       __         .______       __   _______   _______   __       _______
|           ||  |  |  | |   ____|   |   ____||  | |  \ |  |     /   \     |  |        |   _  \     |  | |       \ |       \ |  |     |   ____|
`---|  |----`|  |__|  | |  |__      |  |__   |  | |   \|  |    /  ^  \    |  |        |  |_)  |    |  | |  .--.  ||  .--.  ||  |     |  |__
    |  |     |   __   | |   __|     |   __|  |  | |  . `  |   /  /_\  \   |  |        |      /     |  | |  |  |  ||  |  |  ||  |     |   __|
    |  |     |  |  |  | |  |____    |  |     |  | |  |\   |  /  _____  \  |  `----.   |  |\  \----.|  | |  '--'  ||  '--'  ||  `----.|  |____
    |__|     |__|  |__| |_______|   |__|     |__| |__| \__| /__/     \__\ |_______|   | _| `._____||__| |_______/ |_______/ |_______||_______|
                                                                                                                                                                                                 """)
    print("With a swift move, you free the hostage. They're visibly shaken, but relief washes over their face.")
    time.sleep(2)
    input("\nPress enter to comfort the hostage and ask for any information they might have...")

    print("\nThe hostage takes a deep breath, steadying their nerves. They look at you with determination in their eyes.")
    print("'I overheard something...', they reveal. 'The Zodiac... he mentioned a place...'")
    time.sleep(2)
    input("\nPress enter to encourage the hostage to continue...")

    print("""
\nThe hostage nods, taking another steadying breath before speaking.
"The clue is: 'The Zodiac hides where the sun sets, but never sleeps.'
Possible locations are: cafe, library, museum, park, cinema, gym, train station, west.""")
    time.sleep(2)

    locations = ["cafe", "library", "museum", "park", "cinema", "gym", "train station", "west"]
    for attempt in range(1, 4):
        guess = input(f"\nAttempt {attempt}: Where do you think the Zodiac is? ").lower()

        if guess in locations:
            if guess == "west":
                print("\nFollowing the clue, you head to the west of the city. The setting sun casts long shadows as you approach a seemingly ordinary building.")
                input("\nPress enter to continue your approach...")
                print("\nWith determination, you kick open the door and step inside. The room is quiet and eerie.")
                input("\nPress enter to confront the Zodiac killer...")
                print("\nYou see a figure in the shadows. It's the Zodiac killer. With a swift move, you apprehend him. The case is solved!")
                win()
                return
            else:
                print("You search that location, but the Zodiac killer isn't there.")
        else:
            print("That location doesn't seem to be on your list of possible hideouts.")

        if attempt < 3:
            print("Think about the clue and try again.")
        time.sleep(2)

    print("\nUnfortunately, the Zodiac killer has escaped. You'll need to try again.")
    fail()




if __name__ == "__main__":
    game_start()
```
